```python
#6. In customer table, create a new variable called “customer value segment” using customer value as follows. -  
#- High Value Segment - > 25000  
#- Medium Value Segment – Between 10000 and 25000  
#- Low Value Segment – less than or equal to 10000 
```

```python
def seg(x):
    for x in customer_data.Customer_Value:
        if x>=25000:
            return 'High Value Segment'
        elif x>25000 and x<10000:
            return 'Medium Value Segment'
        else:
            return 'Low Value Segment'
customer_data['customer value segment']= customer_data['Customer_Value'].apply(seg)
customer_data
```

```python
#7. Create variables “average revenue per trip” and “balance points” in the customer data set.  
```

```python
customer_data['average revenue per trip']=(customer_data.Customer_Value*customer_data.Buy_Times)
customer_data['balance points']=(customer_data.Points_Earned - customer_data.Points_Redeemed)
```

```python
#8. How many days between last purchase date and today?
```

```python
customer_data['First_Date']=pd.to_datetime(customer_data.First_Date)
customer_data['Recent_Date']=pd.to_datetime(customer_data.Recent_Date)
customer_data['Days_diff']=customer_data['Recent_Date']-customer_data['First_Date']

customer_data.head(20)
```

```python
#9. Calculate percentage of sales by each last city, state and region.
```

```python
customer_data.groupby(['Customer_ID','Last_city','Last_state','Last_region']).agg({'Customer_Value':'sum'})*100/customer_data.Customer_Value.sum()
```

```python
#10. What is the count of customers, average number of purchases and average purchase transaction value by last state 
#and city. 
```

```python
customer_data.groupby(['Last_state','Last_city']).agg({'Customer_ID':'count','Buy_Times':'mean','Customer_Value':'sum'})
```
